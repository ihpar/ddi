```python
import numpy as np
```

```python
dosya_yollari = [
    "/content/d1.txt",
    "/content/d2.txt",
    "/content/d3.txt"
]

token_2_index = {}
token_index = 0
docs = []

for dosya_yolu in dosya_yollari:
    with open(dosya_yolu, "r") as f:
        icerik = f.readline()
        tokens = icerik.split(" ")
        docs.append(tokens)
        for token in tokens:
            if token not in token_2_index:
                token_2_index[token] = token_index
                token_index += 1

print(token_2_index)
print(docs)
```

```python
sayma_vektorleri = np.zeros((len(dosya_yollari), len(token_2_index)))
```

```python
sayma_vektorleri
```

```python
for i, doc in enumerate(docs):
    for token in doc:
        sayma_vektorleri[i][token_2_index[token]] += 1
```

```python
sayma_vektorleri
```

```python
def cos_similarity(v1, v2):
    return np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))


cos_similarity(np.array([1, 2]), np.array([3, 4]))
```

```python
benzerlik_matrisi = np.zeros((len(dosya_yollari), len(dosya_yollari)))
benzerlik_matrisi
```

```python
for i, doc_vektor_i in enumerate(sayma_vektorleri):
    for j, doc_vektor_j in enumerate(sayma_vektorleri):
        benzerlik = cos_similarity(doc_vektor_i, doc_vektor_j)
        if i == j:
            benzerlik_matrisi[i][j] = -2
        else:
            benzerlik_matrisi[i][j] = benzerlik
```

```python
def en_benzeri_hangisi(doc_index, dosya_yollari):
    """
    doc_index = 0 verilirse -> 0. doc'un (d1.txt) en benzeri hangisi?
    """
    satir = benzerlik_matrisi[doc_index]
    en_benzer_dosya_idx = np.argmax(satir)
    return dosya_yollari[en_benzer_dosya_idx]
```
